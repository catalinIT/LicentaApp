```python
# Import Module
import os
  
# Folder Path
path = r"C:\Users\catat\OneDrive\Desktop\DataForBachelorThesis\Real-lifeDeceptionDetection2016\Transcription\Deceptive"

# Change the directory
os.chdir(path)
  
# Read text File
  
def read_text_file(file_path):
    with open(file_path, 'r') as f:
        print(f.read())
  
  
# iterate through all file
for file in os.listdir():
    # Check whether file is in text format or not
    if file.endswith(".txt"):
        file_path = f"{path}\{file}"
  
        # call read text file function
        read_text_file(file_path)
```

```python
# create a pandas dataframe and populate it with all the records
import pandas as pd
df_transcript = pd.DataFrame(columns=('statement', 'veracity'))

index = 0
# Folder Path
path = r"C:\Users\catat\OneDrive\Desktop\DataForBachelorThesis\Real-lifeDeceptionDetection2016\Transcription\Deceptive"

# Change the directory
os.chdir(path)
  
# Populate the data frame using the line read from the text file
  
def populate_dataframe(file_path, index, df):
    with open(file_path, 'r') as f:
        value_to_add = {'statement': f.read(), 'veracity': 0}
        row_to_add = pd.Series(value_to_add, name=index)
        df.loc[index] = row_to_add
  
  
# iterate through all file
for file in os.listdir():
    # Check whether file is in text format or not
    if file.endswith(".txt"):
        file_path = f"{path}\{file}"
        populate_dataframe(file_path, index, df_transcript)
        index += 1
```
